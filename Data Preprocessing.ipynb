```python
import numpy as np
import pandas as pd
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("car_price_prediction.csv")
```

```python
dataset
```

```python
from sklearn.preprocessing import LabelEncoder

# Copy dataframe
dataset = dataset.copy()

# Columns to encode
categorical_cols = [
    "make", "model", "transmission", "fuel_type", "drivetrain", "body_type",
    "exterior_color", "interior_color", "accident_history", "seller_type",
    "condition", "trim"
]

# Apply LabelEncoder
le = LabelEncoder()
for col in categorical_cols:
    dataset[col] = le.fit_transform(dataset[col].astype(str))
```

```python
dataset
```

```python
#Finding Missing Values
dataset.isnull().sum()
```

```python
dataset.isna().sum()
```

```python
#Finding Quantitative and Qualitative columns
def quanqual(dataset):
    quan=[]
    qual=[]
    for columnname in dataset.columns:
        if(dataset[columnname].dtype=='O'):
            #print("qual")
            qual.append(columnname)
        else:
            #print("quan")
            quan.append(columnname)
    return quan, qual
```

```python
quanqual(dataset)
```

```python
#Save the Preprocessed data 
dataset.to_csv("Preprocessed_car_predict.csv",index=False)
```
